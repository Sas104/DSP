```python
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 2, 1, 3, 2]
m = len(x)  # length of x
n = len(y)  # length of y
r = np.zeros(m + n - 1)  # defining r to store result
f_y = y[::-1]  # folding y

for i in range(1, m + n):
    # padding x with (m-1) number of zeros before and after initial elements
    xp = np.pad(x, (i - 1, m + n - i - 1), mode='constant', constant_values=0)
    # padding y with (n-1) number of zeros before and after initial elements
    yp = np.pad(f_y, (n - 1, n - 1), mode='constant', constant_values=0)
    s = np.multiply(xp, yp)  # multiplying padded arrays
    r[i - 1] = np.sum(s, axis=0)  # assigning results

N = np.arange(0, m + n - 1, 1)
print('r =', r)

# Set up the subplots
plt.figure(figsize=(6, 12))

# Plot x
plt.subplot(3, 1, 1)
plt.stem(np.arange(0, m), x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Signal x')

# Plot y
plt.subplot(3, 1, 2)
plt.stem(np.arange(0, n), y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Signal y')

# Plot the convolution result
plt.subplot(3, 1, 3)
plt.stem(N, r)
plt.xlabel('Time shift, n')
plt.ylabel('Amplitude')
plt.title('Convolution')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```
