```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter, freqz

# Create time vector
sample_rate = 1000  # Sample rate in Hz
duration = 5  # Duration of the signal in seconds
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

# Create three sinusoidal signals
freq_1 = 0.5
freq_2 = 30
freq_3 = 100
signal_1 = np.sin(2 * np.pi * freq_1 * t)
signal_2 = np.sin(2 * np.pi * freq_2 * t)
signal_3 = np.sin(2 * np.pi * freq_3 * t)

# Add the three signals together
sum_signal = signal_1 + signal_2 + signal_3

# Design the Butterworth filter
order = 6  # Filter order
cutoff_freq = 20  # Cutoff frequency in Hz
b, a = butter(order, cutoff_freq / (sample_rate / 2), btype='low', analog=False, output='ba')

# Compute frequency response of the filter
w, h = freqz(b, a, worN=8000)
freq = 0.5 * sample_rate * w / np.pi
db = 20 * np.log10(abs(h))

plt.plot(freq, db)
plt.title('Digital Butterworth Filter Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (dB)')
plt.grid()


plt.figure(figsize=(12, 6))


# Apply the low-pass Butterworth filter
filtered_signal = lfilter(b, a, sum_signal)

# Compute the FFT of the signals
fft_signal = np.fft.fft(sum_signal)
fft_filtered_signal = np.fft.fft(filtered_signal)
freq_fft = np.fft.fftfreq(len(t), d=1.0/sample_rate)

# Plot the original and filtered signals with their FFTs
plt.subplot(2, 2, 1)
plt.plot(t, sum_signal, label='Sum of 0.5Hz, 30Hz, and 100Hz Signals')
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(freq_fft, np.abs(fft_signal), label='FFT of Sum Signal')
plt.title('FFT of Original Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(t, filtered_signal, label='Filtered Signal', linestyle='dashed')
plt.title('Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(freq_fft, np.abs(fft_filtered_signal), label='FFT of Filtered Signal')
plt.title('FFT of Filtered Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import cheby2, lfilter, freqz

# Create time vector
sample_rate = 1000  # Sample rate in Hz
duration = 5  # Duration of the signal in seconds
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

# Create three sinusoidal signals
freq_1 = 0.5
freq_2 = 30
freq_3 = 100
signal_1 = np.sin(2 * np.pi * freq_1 * t)
signal_2 = np.sin(2 * np.pi * freq_2 * t)
signal_3 = np.sin(2 * np.pi * freq_3 * t)

# Add the three signals together
sum_signal = signal_1 + signal_2 + signal_3

# Design the Chebyshev2 band-pass filter
order = 4 # Filter order
low_cutoff_freq = 15  # Low cutoff frequency in Hz
high_cutoff_freq = 40  # High cutoff frequency in Hz
stopband_attenuation = 30  # Stopband attenuation in dB
b, a = cheby2(order, stopband_attenuation, [low_cutoff_freq / (sample_rate / 2), high_cutoff_freq / (sample_rate / 2)], btype='bandpass', analog=False, output='ba')

# Apply the band-pass Chebyshev2 filter
filtered_signal = lfilter(b, a, sum_signal)

# Compute frequency response of the filter
w, h = freqz(b, a, worN=8000)
freq = 0.5 * sample_rate * w / np.pi
db = 20 * np.log10(abs(h))

plt.plot(freq, db)
plt.title('Digital Chebyshev2 Band-Pass Filter Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (dB)')
plt.grid()

# Compute the FFT of the signals
fft_signal = np.fft.fft(sum_signal)
fft_filtered_signal = np.fft.fft(filtered_signal)
freq_fft = np.fft.fftfreq(len(t), d=1.0/sample_rate)


# Plot the original and filtered signals with their FFTs
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(t, sum_signal, label='Sum of 0.5Hz, 30Hz, and 100Hz Signals')
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(freq_fft, np.abs(fft_signal), label='FFT of Sum Signal')
plt.title('FFT of Original Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(t, filtered_signal, label='Filtered Signal', linestyle='dashed')
plt.title('Filtered Signal (Band-Pass Filter)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(freq_fft, np.abs(fft_filtered_signal), label='FFT of Filtered Signal')
plt.title('FFT of Filtered Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
