```python
import numpy as np
import matplotlib.pyplot as plt

f=2             #frequency
t = np.linspace(0, 1, 1000)  
s1 = np.sin(2*np.pi*f*t)  #signal of interest
erms=np.zeros(9)
for n in range(1,9):
    n_bits = 1*n  # number of bits in the quantizer

    q_levels = np.linspace(-1, 1, 2**n_bits)  #quantization level
    q_step = q_levels[1] - q_levels[0]     # step size

    s_q = np.round(s1/q_step)*q_step
    # Quantize the input signal
    error=s_q-s1  #error calculation
    erms[n]=np.sqrt(np.mean(error**2))
    # Plots
    plt.figure(n)
    plt.subplot(2,1,1)
    plt.plot(t, s1, label='Original signal')
    plt.plot(t, s_q, label='Quantized signal')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.legend(loc='upper right')
    plt.title(f"Quantization level={n*2}")
    plt.subplot(2,1,2)
    plt.plot(t, error, label='Error')
    plt.xlabel('Time')
    plt.ylabel('Error Amplitude')
    plt.legend(loc='upper right')
erms=np.delete(erms,0)
n=np.arange(2,17,2)
plt.figure(9)
plt.plot(n,erms,'o-',label='Error RMS')
plt.xlabel('Quantization level')
plt.ylabel('Error RMS')
plt.legend(loc='upper right')
plt.title(f"Quantization level VS RMS Error")
```
